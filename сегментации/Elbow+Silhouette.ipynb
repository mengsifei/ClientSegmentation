```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)
pca = PCA(n_components = 2)
a = pca.fit_transform(scaled)
plt.scatter(a[:, 0], a[:, 1])
pca.explained_variance_ratio_.sum()
```

```python
from sklearn.cluster import KMeans
inertia = []
the_range = range(1, 20)
for k in the_range:
    kmeans = KMeans(n_clusters=k, n_init = 10, random_state = 42).fit(scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(the_range, inertia, marker='s', color='gray');
plt.axvline(5)
plt.axvline(15)
plt.axvline(18)
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title("Elbow method")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_scores = []
for K in range(2, 20):
    model = KMeans(n_clusters=K, n_init=10, random_state=42)
    cluster_assignments = model.fit_predict(scaled)
    silhouette_avg = silhouette_score(scaled, cluster_assignments)
    silhouette_scores.append(silhouette_avg)
```

```python
np.savetxt("silhouette.txt", silhouette_scores)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(2, 20), silhouette_scores, marker='s', color='gray')
plt.axvline(3)
plt.axvline(5)
plt.axvline(15)
plt.axvline(18)
plt.xlabel("Number of clusters: K")
plt.ylabel('Silhouette Score')
plt.title("Silhouette Score")
plt.show()
```
